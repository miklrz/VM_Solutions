```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# f(x) = cos(x) - x
# Найти экстремум функции, точность принять равную ε=0.01

```python
def f(x):
    return np.cos(x) - x
EPS=0.01
y=np.arange(-10,10)
x=np.array([f(i) for i in y])
data = {'x' : x, 'y': y}
df = pd.DataFrame(data)
df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,marker="o", label="f(x)", color="blue")
plt.grid()
```

```python
a = 0
b = 3
while np.abs(a-b) > EPS:
    mid = (a+b)/2
    x1 = mid - EPS
    x2 = mid + EPS
    f1 = f(x1)
    f2 = f(x2)
    if f1 < f2:
        b = x2
    else:
        a = x1
    print(a,b)
```
