```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Данные
x = [14, 15, 17, 19, 20, 21, 22, 24, 27, 32]
y = [70, 69.5, 68.5, 67.5, 66, 65, 64.5, 62.5, 60, 53.5]
```

```python
xi2 = [x[i]**2 for i in range(len(x))]
xi3 = [x[i]**3 for i in range(len(x))]
xi4 = [x[i]**4 for i in range(len(x))]

x2y = [xi2[i]**2*y[i] for i in range(len(x))]
xy = [x[i]*y[i] for i in range(len(x))]
```

```python
df = pd.DataFrame({
    'x': x,
    'y': y,
    'x^2': xi2,
    'x^3': xi3,
    'x^4': xi4,
    'x*y': xy,
    'x^2*y': x2y
})
```

```python
print(df)
print("  ",sum(x), sum(y), sum(xi2), sum(xi3), sum(xi4), sum(xy), sum(x2y))
```

```python
# Построение графика
plt.figure(figsize=(10, 6))
plt.scatter(x, y, marker="o", color="b", label="y vs x")

# Настройки графика
plt.title("График зависимости y от x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```
